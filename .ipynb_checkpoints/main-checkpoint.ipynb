```python
# Already8 is a simple program that allows you to select n number of dishes from your cookbook
# and retrieve a shopping list with the ingredients required to make those dishes.

# import libraries

import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import csv
import random
import operator
from csv import reader
```

```python
data = pd.read_csv('cookbook.csv')

# create a pandas dataframe

data.head()
```

```python
# remove all NaN values and replace with string 'empty'

data = data.replace(np.nan, 'empty')

# make all strings lowercase and remove trailing whitespace and even whitespace at all/substitute an underscore
```

```python
# create a dictionary of the recipes called my_dict from the initial dataframe

cookbook = data.to_dict()

#cookbook ['steak_with_greek_salad'][4] = 'Feta'# adding feta to empty key value in steak with greek salad in cookbook
```

```python
# created the cookbook list manually in Excel because iterating dicts was a pain

cookbook_order = pd.read_csv('cookbook_order_t.csv')
cookbook_order
```

```python
# Create a dict from this ordered dataframe

cookbook_order2 = cookbook_order.to_dict()
```

```python
# FUNCTION 1 - SELECTION IT - VERSION 2

# This function works to take a list of strings which are the dish #, outputs the dish names as strings in a list

def selection_it(items): # pass a list of strings 
    update = []
    for j in items:
        update.append(cookbook_order2[j][0])
    return update
```

```python
# Function 2 - CREATE MENU 

# create_menu() creates the menu dictionary which each dish as a key and the ingredients 
# as a nested dictionary within that

# It does that by getting multiple items from cookbook using the return value of selection_it
# in the form of the second level dictionary with the ingredients (so it will make another dictionary)

def create_menu(selection):
    menu = {}
    for i in range(3):
         menu[selection[i]] = cookbook[selection[i]]
    return menu
```

```python
# FUNCTION 3 - SHOPPING LIST FUNCTION (PREVIOUSLY KNOWN AS "MENU_ITEMS")

# This function provides the shopping list by taking the menu dictionary which was created using selection_it
# and creates a set which contains the unique ingredients required to make those selected dishes (no duplicates).

def shopping_list(menu):
    menu_ingredients = []
    for i in menu.values():
        for j in i.values():
            if j != "empty":
                menu_ingredients.append(j)
    return set(menu_ingredients)
```

```python
# Function 4 - Print Cookbook

# This function just prints the cookbook for the reader
# You can make this better by removing the underscores and using Title Case

def print_cookbook():
    for number, zero in cookbook_order2.items():
        for constant, dish in zero.items():
            print(number, dish)
```

```python
# FUNCTION 5 - MAIN FUNCTION: Already8

# This function presents the user with a visual menu and asks for manual user input through text
# specifically the order number of the dishes required for the shopping list this week

def already_8():
    print_cookbook()
    raw_order = input("Enter dish numbers here, separated by commas and a space, e.g: 4, 5, 6, and then press enter: ")
    order = raw_order.split(", ")
    selection = selection_it(order)
    menu = create_menu(selection)
    return shopping_list(menu)

already_8()
```
